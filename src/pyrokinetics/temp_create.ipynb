```python
import os
import subprocess
from pathlib import Path
import textwrap
import logging
logger = logging.getLogger(__name__)
```

```python

def tglf(run_dir):
    # Make sure the directory exists
    os.makedirs(run_dir, exist_ok=True)

    parent = os.path.dirname(run_dir)
    folder = os.path.basename(run_dir)

    subprocess.run(
        ["tglf", "-e", folder],
        cwd=parent
    )
    logger.info(f"Running simulation in directory: {run_dir}")
```

```python
def tglf_scan(pyro_scan):
    for run_dir in pyro_scan.run_directories:
        tglf(run_dir)
```

```python
import os
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
from pyrokinetics import Pyro, PyroScan, template_dir
```

```python
pyro = Pyro(gk_file = template_dir / "outputs" / "GS2_linear" / "gs2.in")

pyro.gk_code = "TGLF"

# Converted the Code into TGLF

pyro.numerics.nonlinear = True


param_1 = "gamma_exb"
values_1 = (
    np.arange(0, 0.4, 0.1)
    * pyro.norms.pyrokinetics.vref
    / pyro.norms.pyrokinetics.lref
)


# Dictionary of param and values
param_dict = {param_1: values_1}

# Create PyroScan object with more descriptive naming
pyro_scan = PyroScan(
    pyro,
    param_dict,
    value_fmt=".4f",  # Increased precision for small beta values
    value_separator="_",
    parameter_separator="_",
)
# Add proper parameter mapping for beta
pyro_scan.add_parameter_key(
    parameter_key="gamma_exb",
    parameter_attr="numerics",
    parameter_location=["gamma_exb"],
)

# Create scan directory and write input files
pyro_scan.write(
    file_name="input.tglf",
    base_directory= template_dir / "outputs" / "TGLF_transport_scan",
    template_file=None,
    Relative_Path=True,
)
```

```python
tglf_scan(pyro_scan)
```
